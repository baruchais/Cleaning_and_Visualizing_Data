Now lets do some visulization

Depending on what you want to do, there are different types of charts

```python
# Data Visualization 1: Content type in percentage
content_counts = df['type'].value_counts(normalize=True) * 100
plt.figure(figsize=(8, 6))
content_counts.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
plt.title('Content Type Percentage')
plt.ylabel('')
plt.show()
```

```python
# Data Visualization 2: Movie & TV Show by Country (Bar Chart)
country_counts = df['country'].value_counts().head(10)
plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar', color='lightblue')
plt.title('Top 10 Countries with Most Content')
plt.xlabel('Country')
plt.ylabel('Number of Content')
plt.xticks(rotation=45)
plt.show()
```

```python
# Data Visualization 4: Top Directors
top_directors = df['director'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_directors.plot(kind='barh', color='lightgreen')
plt.title('Top 10 Directors with Most Content')
plt.xlabel('Number of Content')
plt.ylabel('Director')
plt.show()
```

```python
# Pair plot
plt.figure(figsize=(8, 6))
pairplot = sns.pairplot(df[['release_year', 'duration']], kind='scatter')

# Adjust x-axis limits for release year
pairplot.axes[0, 0].set_xlim(2000, 2025)

plt.suptitle('Pair Plot of Release Year and Duration', y=1.2)
plt.show()
```

# What is feature engineering?

Feature engineering is like choosing the right ingredients for a recipe. Just like you pick the best ingredients to make a dish tasty, in machine learning, you pick the right information (or ‚Äúfeatures‚Äù) from your data to help your model make good predictions.

Sometimes, some ingredients (or ‚Äúfeatures‚Äù) don‚Äôt add any flavor to your dish (or ‚Äúmodel‚Äù), so you leave them out. Other times, two ingredients might have the same flavor, so you only need one of them. This is what we do in feature engineering - we keep the useful features and remove the unnecessary ones to make our model work better.

In short, feature engineering is all about selecting the best ‚Äúingredients‚Äù for your machine learning ‚Äúrecipe‚Äù.

```python
from sklearn.preprocessing import LabelEncoder

# creating instance of labelencoder
labelencoder = LabelEncoder()

# Assigning numerical values and storing in another column
df['type_encoded'] = labelencoder.fit_transform(df['type'])
df['rating_encoded'] = labelencoder.fit_transform(df['rating'])

# Now you can create a correlation matrix
corr_matrix = df[['type_encoded', 'rating_encoded']].corr()

# And create a heatmap of this correlation matrix
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

This tells us that the correlation is not that strong (0.11 is closer to 0).
