## Handling Missing Values

Generally, there are 2 ways to deal with missing values:
- Remove them completely
- Or replace them with an approximate

How do you know which one to choose?
- If you have large dataset with thousands of values then it OK to simply remove the missing values
- Elsewise we can impute the value with either the mode or mean depending on the type of the data

What are the 2 types of values?
- Discrete Values: countable (i.e. number of students in a classroom, flipping a coin)
    - It is IMPOSSIBLE to have 2.5 students
    - NOTE: Discrete Values can be both numerical or categorial data
- Continuous Values: measurable (i.e the height of students)
    - It is possible for a student to have the height of 5.75 feet

Imputing Missing Values:
We impute missing value depending on the type of value:
- If its a Discrete Value, we must impute it using its mode value (Note: Sometimes you may need to convert it into numerical value before finding the mode)
- If its a Continious Value, we must impute it using its mean value

For the sake of this exercise lets impute its missing values:

```python
df.isna().sum()
```

```python
df[df['Teacher_Quality'].isnull()]
```

```python
# Firstly we need to check if its discrete or continuous data
df['Teacher_Quality'].unique()
```

```python
tq_mode_value = df['Teacher_Quality'].mode()[0] # if theres a tie it'll pic an arbitary one
```

```python
df['Teacher_Quality'] = df['Teacher_Quality'].fillna(tq_mode_value)
```

```python
df['Parental_Education_Level'].unique()
```

```python
pel_mode_value = df['Parental_Education_Level'].mode()[0] # if theres a tie it'll pic an arbitary one
```

```python
df['Parental_Education_Level'] = df['Parental_Education_Level'].fillna(pel_mode_value)
```

```python
df['Distance_from_Home'].unique()
```

```python
dfm_mode_value = df['Distance_from_Home'].mode()[0] # if theres a tie it'll pic an arbitary one
```

```python
df['Distance_from_Home'] = df['Distance_from_Home'].fillna(dfm_mode_value)
```
